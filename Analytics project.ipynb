# Hosts vs Super hosts

```python
import matplotlib.pyplot as plt

# Create separate dataframes for superhosts and hosts
superhosts_data = df[df['Superhost'] == True]
hosts_data = df[df['Superhost'] == False]

# Check if there are data points for superhosts and hosts
if len(superhosts_data) > 0 and len(hosts_data) > 0:
    # Plot the distributions of person capacity for superhosts and hosts
    plt.figure(figsize=(8, 6))
    plt.hist(superhosts_data['Person Capacity'], bins=10, alpha=0.5, label='Superhosts')
    plt.hist(hosts_data['Person Capacity'], bins=10, alpha=0.5, label='Hosts')
    plt.xlabel('Person Capacity')
    plt.ylabel('Frequency')
    plt.title('Distribution of Person Capacity for Superhosts and Hosts')
    plt.legend()
    plt.show()
else:
    print("No data points for superhosts or hosts.")
```

```python
# Calculate the average person capacity for superhosts and hosts
superhosts_avg_rooms = df[df['Superhost'] == True]['Multiple Rooms'].mean()
hosts_avg_rooms = df[df['Superhost'] == False]['Multiple Rooms'].mean()

# Create data for the bar graph
labels = ['Superhosts', 'Hosts']
sizes = [superhosts_avg_rooms, hosts_avg_rooms]
colors = ['green', 'orange']

# Create the bar graph
plt.figure(figsize=(6, 6))
plt.bar(labels, sizes, color=colors)
plt.title('Average Person Capacity Comparison: Superhosts vs. Hosts', fontweight='bold', color='red')
plt.xlabel('Host Type')
plt.ylabel('Average Person Capacity')
plt.show()
```

>Super hosts generally provide a greater person capacity than normal hosts. I would recommend the hosts specially in areas like Barcelona, Amsterdam, Berlin and Paris to start increasing the person capacity as this would help them cater to single travelling customers as well as customers who travel in herd or with their families.

# Distribution of hosts and super hosts

>It seems that there is a paucity of Airbnb super hosts as out of all the hosts, they only constitute for 28 percent of the overall proportion. This roughly means that one out of every 4 host is a super host.

>Also if we look at the city wise super hosts, the distribution is highly sporadic as more than 50 % of the super hosts are concentrated in three cities namely Rome, Athens and Budapest out of nine!!!
